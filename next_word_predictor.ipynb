```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer()

tokenizer.fit_on_texts([faqs])
```

```python
tokenizer.word_index
```

```python
l=[]
for sentence in faqs.split('.'):
  tokenized_sentences=tokenizer.texts_to_sequences([sentence])[0]
  for i in range(1,len(tokenized_sentences)):
    l.append(tokenized_sentences[:i+1])
```

```python
max_len=max([len(x) for x in l])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_sequences=pad_sequences(l,maxlen=max_len,padding='pre')
```

```python
x=padded_sequences[:,:-1]
y=padded_sequences[:,-1]
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=len(tokenizer.word_index)+1)
```

```python
y.shape
```

```python
from keras.models import Sequential
from keras.layers import Embedding,LSTM,Dense,Dropout
```

```python
model=Sequential()
```

```python
model.add(Embedding(len(tokenizer.word_index)+1,100,input_length=max_len))
model.add(LSTM(150))
model.add(Dropout(0.2))
model.add(Dense(len(tokenizer.word_index)+1,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(x,y,epochs=100)
```

```python
model.save('model.h5')
```

```python
import numpy as np
text="marvel"
for i in range(10):
  token_text=tokenizer.texts_to_sequences([text])[0]
  p=pad_sequences([token_text],maxlen=max_len,padding='pre')

  pos=np.argmax(model.predict(p))
  for word,index in tokenizer.word_index.items():
    if index==pos:
      text=text+" "+word
      print(text)
```
